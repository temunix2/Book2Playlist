```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
guten_books = pd.read_pickle('books_content.pkl')
```

```python
text = []
```

```python
text = list(guten_books['content'])
```

```python
stop = stopwords.words('english')
stop += ['.', ',', '(', ')', "'", '"']
stop = set(stop)
```

## Count Vectorizer

```python
vectorizer = CountVectorizer(stop_words='english',min_df = 0.01)
vectorizer.fit(text)
```

```python
print(len(vectorizer.vocabulary_))
```

```python
print(vectorizer.get_feature_names()[0:30])
```

```python
# encode document
vector = vectorizer.transform(text)
```

```python
# summarize encoded vector
print(vector.shape)
```

```python
print(type(vector))
x_back = vector.toarray()
print(x_back)
```

```python
test = pd.DataFrame(x_back, columns=vectorizer.get_feature_names())
```

## TF-IDF

```python
vectorizer = TfidfVectorizer(min_df=0.01)
vectorizer.fit(text)
```

```python
# summarize
print(len(vectorizer.vocabulary_))
print(len(vectorizer.idf_))
```

```python
# encode document
vector = vectorizer.transform(text)
```

```python
# summarize encoded vector
print(vector.shape)
xf_back = vector.toarray()
print(xf_back)
```

# Count Vectorizer Curve 

```python
count = {}
for i in np.arange(0.0,0.01,0.002):
    vectorizer = CountVectorizer(stop_words='english',min_df = i)
    vectorizer.fit(text)
    count[i] = len(vectorizer.vocabulary_)
```

```python
count
```

```python
keys = []
val = []
for word, meaning in count.items():
    keys.append(word)
    val.append(meaning)
```

regex for tokenizer \\b[a-z][a-z]+\\b

```python
plt.plot(keys,val)
```

```python
lemma=nltk.stem.WordNetLemmatizer()
```

```python
# pip install textblob
from textblob import TextBlob
```

```python
lem_text = []
for num in range(len(text[0:10])):
    temp = ""
    for word in TextBlob(text[num]).words:
        temp += lemma.lemmatize(word)
        temp += ' '
    lem_text.append(temp)
```
